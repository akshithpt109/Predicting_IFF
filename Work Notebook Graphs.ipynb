```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("Corruption Dataset (2009 to 2018) vFinal - W Names.csv")
```

```python
df.head()
```

```python
df.set_index(df["Country"],inplace=True)
df.drop("Country",axis=1,inplace=True)
```

```python
ax = sns.violinplot(data=df, x="IFF")
# Set the y-axis title
ax.set_ylabel('Density')
ax.set_xlabel('IFF (Millions of USD)')
# Show the plot
plt.show()
```

```python
ax = sns.violinplot(data=df,x='Aid Percentage of Gross National Income')
# Set the y-axis title
ax.set_ylabel('Density')
ax.set_xlabel('Aid Percentage of Gross National Income')
# Show the plot
plt.show()
```

```python
ax = sns.violinplot(data=df,x='Population')
# Set the y-axis title
ax.set_ylabel('Density')
ax.set_xlabel('Population (Billions)')
# Show the plot
plt.show()
```
